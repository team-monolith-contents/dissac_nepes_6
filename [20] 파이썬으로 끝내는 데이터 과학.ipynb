###  **1. Í∑∏ÎûòÌîÑÏùò Î∞∞Í≤Ω Íæ∏ÎØ∏Í∏∞**

> Í∑∏ÎûòÌîÑÎ•º Í∑∏Î¶¨Îäî Ïó¨Îü¨ Ìï®ÏàòÏóêÎäî Ïó¨Îü¨Í∞ÄÏßÄÏùò Î∞∞Í≤Ω Íæ∏ÎØ∏Í∏∞ Í∏∞Îä•Ïù¥ Ïà®Ïñ¥ÏûàÏäµÎãàÎã§.  
> Îã§Ïùå ÏΩîÎìúÎ•º Ïã§ÌñâÌïòÏó¨ Í∑∏ÎûòÌîÑÎ•º ÏñªÏñ¥ ÌôïÏù∏Ìï¥Î≥¥Í≥† Ïñ¥Îñ§ Í∏∞Îä•Ïù¥ ÏÇ¨Ïö©ÎêêÎäîÏßÄ Ï∞æÏïÑÎ¥ÖÏãúÎã§.  


```python
# ÏòàÏ†ú 1: Îã§ÏñëÌïú Î∞∞Í≤Ω Íæ∏ÎØ∏Í∏∞ Í∏∞Îä•Ïù¥ ÏÇ¨Ïö©Îêú Í∑∏ÎûòÌîÑ ÌôïÏù∏ÌïòÍ∏∞

import numpy as np

plt.figure(figsize=(10, 7), facecolor='pink', edgecolor='c', linewidth=2)
plt.axes(facecolor='y')

plt.title("ÏôºÏ™Ω Ï†úÎ™©", loc='left')
plt.title("Ïò§Î•∏Ï™Ω Ï†úÎ™©", loc='right')
plt.title("Ï†úÎ™©, fontsize=40", fontsize=40, fontweight='bold', pad=20)

plt.xlabel("xÏ∂ï Ïù¥Î¶Ñ, fontsize=20", fontsize=20, loc='right')
plt.ylabel("yÏ∂ï Ïù¥Î¶Ñ", rotation='horizontal')

plt.xticks([-2, -1, 0, 1, 2], fontsize=15, fontweight='bold', color='purple')
plt.yticks([np.exp(-2), np.exp(-1), np.exp(0), np.exp(1), np.exp(2)], 
           ["exp(-2)="+str(np.exp(-2)), "exp(-1)="+str(np.exp(-1)), "exp(0)="+str(np.exp(0)), "exp(1)="+str(np.exp(1)), "exp(2)="+str(np.exp(2))]
           ,rotation=30)

plt.xlim(-2.5, 2.5)
plt.ylim(0, 8)

plt.grid(axis='x', color='purple', linestyle='--', linewidth=3)
plt.grid(axis='y', color='gray')

plt.show()
```

> * **axes**: Ï¢åÌëúÏ∂ïÏùÑ ÏùòÎØ∏. ÎààÍ∏àÍ≥º Í≤©Ïûê, Í∑∏ÎûòÌîÑÎ•º Ìè¨Ìï®.

### **3. Í∑∏ÎûòÌîÑ Ï†úÎ™©Í≥º Ï∂ïÏùò Ïù¥Î¶Ñ ÏÑ§Ï†ïÌïòÍ∏∞**

> FigureÎäî ÎëêÍ∞ÄÏßÄ Ï¢ÖÎ•òÏùò Í∏ÄÏûêÎ•º Ìè¨Ìï®Ìï©ÎãàÎã§.  
>
> * **title**: Í∑∏ÎûòÌîÑÏùò Ï†úÎ™©
> * **xlabel/ylabel**: x/yÏ∂ïÏùò Ïù¥Î¶Ñ
> 
> Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∑∏ÎûòÌîÑ Ï†úÎ™©, x/yÏ∂ï Ïù¥Î¶ÑÏùò ÏúÑÏπò, ÌÅ¨Í∏∞, Í∏ÄÏûê Ïä§ÌÉÄÏùºÏùÑ ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.  
>
> * `plt.title("Ï†úÎ™©", fontsize=N, color=c, fontweight=weight, pad=M, loc=lacation)`  
> : "Ï†úÎ™©"Ïù¥ÎùºÎäî Ï†úÎ™©ÏùÑ Í∏ÄÏûêÌÅ¨Í∏∞ N(Ïà´Ïûê), Í∏ÄÏûêÏÉâÏÉÅÏùÄ c(ÏÉâÍπîÏù¥Î¶Ñ), Í∏ÄÏûêÍµµÍ∏∞Î•º weight(Ïà´Ïûê ÌòπÏùÄ **normal/bold**)Î°ú ÏÑ§Ï†ï  
> : Ï†úÎ™©Í≥º axes ÏÇ¨Ïù¥Ïùò Í∞ÑÍ≤©ÏùÑ M(Ïà´Ïûê)Ìè¨Ïù∏Ìä∏Î°ú ÏÑ§Ï†ï  
> : Ï†úÎ™©Ïùò ÏúÑÏπòÎ•º location(**left(Ï¢å)/center(Ï§ëÏïô)/right(Ïö∞)**)ÏúºÎ°ú ÏÑ§Ï†ï  
>
> * `plt.xlabel("xÏ∂ï", fontsize=P, color=c, fontweight=weight, loc=location, rotation=direction)`  
> : "xÏ∂ï"Ïù¥ÎùºÎäî Ï∂ïÏù¥Î¶ÑÏùÑ Í∏ÄÏûêÌÅ¨Í∏∞ P(Ïà´Ïûê), Í∏ÄÏûêÏÉâÏÉÅÏùÄ c(ÏÉâÍπîÏù¥Î¶Ñ), Í∏ÄÏûêÍµµÍ∏∞Î•º weight(Ïà´Ïûê ÌòπÏùÄ **normal/bold**)Î°ú ÏÑ§Ï†ï  
> : Ï∂ïÏù¥Î¶ÑÏùò ÏúÑÏπòÎ•º location(**left(Ï¢å)/center(Ï§ëÏïô)/right(Ïö∞)**)ÏúºÎ°ú ÏÑ§Ï†ï  
> : Ï∂ïÏù¥Î¶ÑÏùò Í∏ÄÏî®Î∞©Ìñ•ÏùÑ direction(**horizontal(Í∞ÄÎ°úÎ∞©Ìñ•)/vertical(ÏÑ∏Î°úÎ∞©Ìñ•)**)ÏúºÎ°ú ÏÑ§Ï†ï  
>
> * `plt.ylabel("yÏ∂ï", fontsize=P, color=c, fontweight=weight, loc=location, rotation=direction)`  
> : loc Ïù∏Ïàò Ïô∏ÏóêÎäî xlabelÍ≥º Í∞ôÏùå  
> : Ï∂ïÏù¥Î¶ÑÏùò ÏúÑÏπòÎ•º location(**top(ÏúÑ)/center(Ï§ëÏïô)/bottom(ÏïÑÎûò)**)ÏúºÎ°ú ÏÑ§Ï†ï  


```python
# ÏòàÏ†ú 3: Í∑∏ÎûòÌîÑ Ï†úÎ™©Í≥º Ï∂ïÏù¥Î¶ÑÏùò ÎÇ¥Ïö©, ÌÅ¨Í∏∞, Ïä§ÌÉÄÏùº, ÏúÑÏπòÎ•º Í≤∞Ï†ï

## (1) Í∑∏ÎûòÌîÑ Ï†úÎ™© ÏÑ§Ï†ï
# "ÏôºÏ™Ω Ï†úÎ™©"ÏùÑ ÏôºÏ™ΩÏóê Í∏ÄÏûêÌÅ¨Í∏∞ 20, Îπ®Í∞ÑÏÉâÏúºÎ°ú Ï∂îÍ∞Ä
plt.title("ÏôºÏ™Ω Ï†úÎ™©", loc='left', fontsize=20, color='r')

# "Ïò§Î•∏Ï™Ω Ï†úÎ™©"ÏùÑ Ïò§Î•∏Ï™ΩÏóê Í∏ÄÏûêÌÅ¨Í∏∞ 20, normal ÍµµÍ∏∞Î°ú Ï∂îÍ∞Ä
plt.title("Ïò§Î•∏Ï™Ω Ï†úÎ™©", loc='right', fontsize=20, fontweight='normal')  

# "Ï§ëÏïô Ï†úÎ™©"ÏùÑ Ï§ëÏïôÏóê Í∏ÄÏûêÌÅ¨Í∏∞ 20, bold ÍµµÍ∏∞Î°ú Ï∂îÍ∞Ä ÌõÑ axesÏôÄÏùò Í∞ÑÍ≤©ÏùÑ 20Ìè¨Ïù∏Ìä∏Î°ú ÏÑ§Ï†ï
plt.title("Ï§ëÏïô Ï†úÎ™©", fontsize=20, fontweight='bold', pad=20)  


## (2) Í∑∏ÎûòÌîÑ Ï∂ï Ïù¥Î¶Ñ ÏÑ§Ï†ï
# xÏ∂ïÏùò Ïù¥Î¶Ñ "xÏ∂ï Ïù¥Î¶Ñ"ÏùÑ Ïò§Î•∏Ï™ΩÏóê Í∏ÄÏûêÌÅ¨Í∏∞ 15, ÌååÎûÄÏÉâ, bold ÍµµÍ∏∞Î°ú Ï∂îÍ∞Ä
plt.xlabel("xÏ∂ï Ïù¥Î¶Ñ", loc='right', fontsize=15, color='b', fontweight='bold')  

# yÏ∂ïÏùò Ïù¥Î¶Ñ "yÏ∂ï Ïù¥Î¶Ñ"ÏùÑ ÏúÑÏ™ΩÏóê Í∏ÄÏûêÌÅ¨Í∏∞ 15, bold ÍµµÍ∏∞Î°ú Í∞ÄÎ°úÎ∞©Ìñ• Í∏ÄÏûêÎ°ú Ï∂îÍ∞Ä
plt.ylabel("yÏ∂ï Ïù¥Î¶Ñ", loc='top', fontsize=15, fontweight='bold', rotation='horizontal')  

plt.show()  # figureÎ•º Ï∂úÎ†•
```

### **4. Ï∂ïÏùò ÎààÍ∏à ÏÑ§Ï†ïÌïòÍ∏∞**

> AxesÎäî ÏïÑÎûòÏóê xÏ∂ï, ÏôºÏ™ΩÏóê yÏ∂ïÏùÑ Í∞ÄÏßëÎãàÎã§.   
>
> * **xticks/yticks**: x/yÏ∂ïÏùò ÎààÍ∏à Ïù¥Î¶Ñ
> * **xlim/ylim**: x/yÏ∂ïÏùò Î≤îÏúÑ
> 
> Îã§ÏùåÍ≥º Í∞ôÏù¥ Í∞Å Ï∂ïÏùò ÎààÍ∏àÏùÑ ÏÑ§Ï†ïÌï† Ïàò ÏûàÍ≥†, Î≥¥Ïó¨ÏßÄÎäî Ï∂ïÏùò Î≤îÏúÑÎ•º ÏÑ§Ï†ïÌï† Ïàò ÏûàÏäµÎãàÎã§.  
>
> * `plt.xticks/yticks(num_data, label_data, fontsize=N, color=c, fontweight=weight, rotation=angle)`  
> : num_dataÏùò Ïà´ÏûêÎì§Ïùò Í∞ÑÍ≤©Ïóê ÎßûÏ∂îÏñ¥ ÎààÍ∏àÏùÑ ÏßÄÏ†ï  
> : label_dataÍ∞Ä ÏûàÎäî Í≤ΩÏö∞ ÏúÑÏùò ÎààÍ∏à Ïù¥Î¶ÑÏùÑ label_dataÎ°ú ÎåÄÏ≤¥    
> : ÎààÍ∏à Ïù¥Î¶ÑÏùÑ Í∏ÄÏûêÌÅ¨Í∏∞ N(Ïà´Ïûê), Í∏ÄÏûêÏÉâÏÉÅÏùÄ c(ÏÉâÍπîÏù¥Î¶Ñ), Í∏ÄÏûêÍµµÍ∏∞Î•º weight(Ïà´Ïûê ÌòπÏùÄ normal/bold)Î°ú ÏÑ§Ï†ï  
> : ÎààÍ∏à Ïù¥Î¶ÑÏùÑ angle(Ïà´Ïûê) Í∞ÅÎèÑÎßåÌÅº Î∞òÏãúÍ≥ÑÎ∞©Ìñ•ÏúºÎ°ú ÌöåÏ†Ñ  
>
> * `plt.xlim/ylim(minumum, maximum)`  
> : axesÏóê Î≥¥Ïó¨ÏßÄÎäî Ï∂ïÏùò Î≤îÏúÑÎ•º minumumÎ∂ÄÌÑ∞ maximumÍπåÏßÄÎ°ú ÏÑ§Ï†ï 

```python
# ÏòàÏ†ú 4: Ï∂ïÏùò ÎààÍ∏à ÏúÑÏπòÏôÄ Ïù¥Î¶ÑÏùÑ Í≤∞Ï†ï

## (1) Ï∂ïÏùò ÎààÍ∏à ÏÑ§Ï†ï
# xÏ∂ïÏùò ÎààÍ∏àÏùÑ ÏùºÏ†ïÌïú Í∞ÑÍ≤©ÏúºÎ°ú ÏßÄÏ†ïÌïòÍ≥† ÎààÍ∏à Ïù¥Î¶ÑÏùÑ Í∏ÄÏûêÌÅ¨Í∏∞ 15, boldÍµµÍ∏∞, Î≥¥ÎùΩÏÉâÏúºÎ°ú Ï∂îÍ∞Ä
plt.xticks([-2, -1, 0, 1, 2], fontsize=15, fontweight='bold', color='purple')  

# yÏ∂ïÏùò ÎààÍ∏àÏùÑ Îã§ÏùåÍ≥º Í∞ôÏù¥ Ï∂îÍ∞Ä
plt.yticks([np.exp(-2), np.exp(-1), np.exp(0), np.exp(1), np.exp(2)],  # yÏ∂ïÏùò ÎààÍ∏àÏùÑ exponential Ìï®ÏàòÎ•º Ïù¥Ïö©Ìï¥ Îã§ÏñëÌïú Í∞ÑÍ≤©ÏúºÎ°ú ÏßÄÏ†ï
           ["exp(-2)="+str(np.exp(-2)), "exp(-1)="+str(np.exp(-1)), "exp(0)="+str(np.exp(0)), "exp(1)="+str(np.exp(1)), "exp(2)="+str(np.exp(2))]  # ÎààÍ∏à Ïù¥Î¶ÑÏùÑ 'exp(n)=Ïã§Ï†úÍ∞í'ÏúºÎ°ú ÏßÄÏ†ï
           ,rotation=30)  # ÎààÍ∏à Ïù¥Î¶ÑÏùÑ Î∞òÏãúÍ≥ÑÎ∞©Ìñ•ÏúºÎ°ú 30ÎèÑ ÌöåÏ†Ñ


## (2) Ï∂ïÏùò Î≤îÏúÑ ÏÑ§Ï†ï
plt.xlim(-2.5, 2.5)  # xÏ∂ïÏùÑ -2.5Î∂ÄÌÑ∞ 2.5ÍπåÏßÄ Î≥¥Ïó¨ÏßÄÎèÑÎ°ù ÏÑ§Ï†ï
plt.ylim(0, 8)  # yÏ∂ïÏùÑ 0Î∂ÄÌÑ∞ 8ÍπåÏßÄ Î≥¥Ïó¨ÏßÄÎèÑÎ°ù ÏÑ§Ï†ï

plt.show()  # figureÎ•º Ï∂úÎ†•
```

```python
# ÏòàÏ†ú 5: figureÏôÄ Ïó¨Îü¨ axes Íæ∏ÎØ∏Í∏∞

## (1) figure ÏÉùÏÑ±
# 3Í∞úÏùò axesÍ∞Ä Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 16Ïù∏Ïπò ÎÜíÏù¥ 6Ïù∏Ïπò ÍµµÍ∏∞ 3, Í≤ÄÏ†ïÏÉâ Î™®ÏÑúÎ¶¨Ïùò figureÎ•º Î∞òÌôò
fig, axes = plt.subplots(1, 3, figsize=(16,6), edgecolor='k', linewidth=3)

## (2) figure Ï†úÎ™© ÏÑ§Ï†ï
fig.suptitle('Ïó¨Îü¨ Í∑∏ÎûòÌîÑ Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞', fontsize=20, fontweight='bold')

# ÏõêÍ∑∏ÎûòÌîÑ ÌäπÏßï ÏÑ§Ï†ï
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 1}
# ÏõêÏùò Ï§ëÏã¨ÏúºÎ°ú Î∂ÄÌÑ∞ Î∞òÏßÄÎ¶ÑÏùò 30%Ïóê Ìï¥ÎãπÌïòÎäî Î∂ÄÎ∂ÑÏùÑ ÏûòÎùºÎÉÖÎãàÎã§. 
# Î∂ÄÏ±ÑÍº¥Ïóê ÎëêÍªòÍ∞Ä 1pixelÏù¥Í≥† ÏÉâÏÉÅÏùÄ ÌïòÏñÄ Î™®ÏÑúÎ¶¨ ÏÑ†ÏùÑ Í∑∏Î¶ΩÎãàÎã§.

## (3) axes[0] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
AÎ∞òÌõÑÎ≥¥ = ['1Î≤à', '2Î≤à', '3Î≤à', '4Î≤à', '5Î≤à', '6Î≤à', '7Î≤à']
AÎ∞òÎìùÌëú = [4, 2, 1, 9, 8, 0, 1]
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[0].pie(AÎ∞òÎìùÌëú, labels=AÎ∞òÌõÑÎ≥¥, autopct='%.1f', wedgeprops=wedgeprops)  # axes[0]Ïóê ÏõêÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Ï†úÎ™© ÏÑ§Ï†ï
axes[0].set_title("AÎ∞ò", fontsize=20) 

## (4) axes[1] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
BÎ∞òÌõÑÎ≥¥ = ['1Î≤à', '2Î≤à', '3Î≤à', '4Î≤à', '5Î≤à']
BÎ∞òÎìùÌëú = [14, 4, 2, 2, 3]
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[1].pie(BÎ∞òÎìùÌëú, labels=BÎ∞òÌõÑÎ≥¥, autopct='%.1f', wedgeprops=wedgeprops)  # axes[1]Ïóê ÏõêÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Ï†úÎ™© ÏÑ§Ï†ï
axes[1].set_title("BÎ∞ò", fontsize=20) 

## (5) axes[2] Í∑∏ÎûòÌîÑ ÏÑ§Ï†ï
# Îç∞Ïù¥ÌÑ∞
CÎ∞òÌõÑÎ≥¥ = ['1Î≤à', '2Î≤à', '3Î≤à', '4Î≤à', '5Î≤à', '6Î≤à']
CÎ∞òÎìùÌëú = [1, 6, 0, 9, 7, 2]
# Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
axes[2].pie(CÎ∞òÎìùÌëú, labels=CÎ∞òÌõÑÎ≥¥, autopct='%.1f', wedgeprops=wedgeprops)  # axes[2]Ïóê ÏõêÍ∑∏ÎûòÌîÑÎ•º Í∑∏Î¶º
# Ï†úÎ™© ÏÑ§Ï†ï
axes[2].set_title("CÎ∞ò", fontsize=20) 

plt.show()  # figureÎ•º Ï∂úÎ†•
```

### **3. Í∑∏ÎûòÌîÑ 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞**


**(1) subplotsÌï®ÏàòÍ∞Ä Î∞òÌôòÌïòÎäî figureÏôÄ axes Î¶¨Ïä§Ìä∏**

> Îã§ÏùåÍ≥º Í∞ôÏù¥ `subplots` Ìï®ÏàòÏóê figureÏóê Ï∂îÍ∞ÄÌï† axesÏóê Í∞úÏàòÎ•º ÎÑ£Ïñ¥ 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥Îêú axesÏôÄ figureÎ•º ÏñªÏùÑ Ïàò ÏûàÏäµÎãàÎã§.  
>
> `figure, axes = plt.subplots(N, M, figsize=(w, h))`   
> : ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú 2Í∞ú, Í∞ÄÎ°úÎ∞©Ìñ•ÏúºÎ°ú 3Í∞ú Ï¥ù 6Í∞úÏùò axesÍ∞Ä ÎÇòÏó¥Îêú ÎÑàÎπÑÍ∞Ä w Ïù∏Ïπò, ÎÜíÏù¥Í∞Ä h Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò  
>  
> ÏúÑÏóêÏÑú ÏñªÏùÄ axesÎäî Ïó¨Îü¨Í∞úÏùò Í∑∏ÎûòÌîÑ(axes)Î•º Ìè¨Ìï®ÌïòÎäî Î¶¨Ïä§Ìä∏ÏûÖÎãàÎã§.  
> Îî∞ÎùºÏÑú Í∞Å Í∑∏ÎûòÌîÑ(axes)Î•º Íæ∏Î∞Ä ÎïåÏóêÎäî `axes[i, j]` ÏôÄ Í∞ôÏù¥ ÏàúÏÑúÎ•º 2Ï∞®Ïõê Ï¢åÌëúÎ•º Ïù¥Ïö©Ìï¥ Ïù∏Îç±Ïã±ÌïòÏó¨ Î∂àÎü¨ÏôÄ ÏÇ¨Ïö©Ìï©ÎãàÎã§.  
>
> Îã§Ïùå ÏòàÏ†úÎ•º Ïã§ÌñâÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑ(axes)Ïùò Î≤àÌò∏Î•º ÌôïÏù∏ÌïòÏó¨ Î¥ÖÏãúÎã§. 

```python
# ÏòàÏ†ú 6: Ïó¨Îü¨ axesÍ∞Ä 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥Îêú figure ÌôïÏù∏ÌïòÍ∏∞

fig, axes = plt.subplots(2, 3, figsize=(10, 6))  # 2Í∞úÏùò axesÍ∞Ä ÏÑ∏Î°úÎ∞©Ìñ•ÏúºÎ°ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 10Ïù∏Ïπò ÎÜíÏù¥ 6Ïù∏ÏπòÏù∏ figureÏôÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò

fig.suptitle("Í∑∏ÎûòÌîÑ Ïó¨Îü¨Í∞ú ÎÇòÏó¥ÌïòÍ∏∞", fontsize=20)

for i in range(2):
    for j in range(3):
        axes[i, j].set_title("axes["+str(i)+", "+str(j)+"]")
plt.show()
```

**(2) figureÏôÄ Ïó¨Îü¨Í∞ú axes Íæ∏ÎØ∏Í∏∞**

> 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥Îêú Í∑∏ÎûòÌîÑÎ•º Íæ∏ÎØ∏Îäî Î∞©Î≤ïÎèÑ Îã§Î•¥ÏßÄ ÏïäÏäµÎãàÎã§.  
>
> Ïù¥Î≤àÏóêÎäî forÎ¨∏ÏùÑ Ïù¥Ïö©ÌïòÏó¨ Í∞Å Í∑∏ÎûòÌîÑÏóê Ï†ëÍ∑ºÌïòÍ≥† ÌûàÏä§ÌÜ†Í∑∏Îû®ÏùÑ Í∑∏Î†§Î≥¥Í≤†ÏäµÎãàÎã§. 

```python
# ÏòàÏ†ú 7: figureÏôÄ Ïó¨Îü¨ axes Íæ∏ÎØ∏Í∏∞, forÎ¨∏ Ïù¥Ïö©ÌïòÍ∏∞

## (1) figure ÏÉùÏÑ±
# axesÍ∞Ä Í∞ÄÎ°úÎ∞©Ìñ• 2Ï§Ñ ÏÑ∏Î°úÎ∞©Ìñ• 3Ï§ÑÎ°ú Ï¥ù 6Í∞ú ÎÇòÏó¥ÎêòÏñ¥ ÏûàÍ≥†, ÎÑàÎπÑ 16Ïù∏Ïπò ÎÜíÏù¥ 10Ïù∏ÏπòÏùò figureÎ•º Î∞òÌôò
# ÎòêÌïú xÏ∂ïÍ≥º yÏ∂ïÏù¥ Í∞ôÏùÄ axesÎ¶¨Ïä§Ìä∏Î•º Î∞òÌôò
fig, axes = plt.subplots(2, 3, figsize=(16, 10), sharex=True, sharey=True)

## (2) figure Ï†úÎ™© ÏÑ§Ï†ï
fig.suptitle('Ïó¨Îü¨ Í∑∏ÎûòÌîÑ 2Ï∞®ÏõêÏúºÎ°ú ÎÇòÏó¥ÌïòÍ∏∞', fontsize=20, fontweight='bold')

## (3) Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ Î∞è Íæ∏ÎØ∏Í∏∞
for i in range(2):
    for j in range(3):
        # Îç∞Ïù¥ÌÑ∞
        numbers = i*10 + 3*(j+1)*np.random.standard_normal(10000)  # ÌëúÏ§ÄÏ†ïÍ∑úÎ∂ÑÌè¨Î•º ÌôïÎ•†Î∞ÄÎèÑÌï®ÏàòÎ°ú ÌïòÏó¨ ÎÇúÏàòÎ•º 10000Í∞ú ÏÉùÏÑ±ÌïòÍ≥† Ï¢åÌëúÏóê Îî∞Îùº ÌäπÏ†ïÌïú ÏàòÎ•º Í≥±ÌïòÍ±∞ÎÇò ÎçîÌï®
        # Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
        axes[i, j].hist(numbers, bins=100, density=True, color='lime')  # Í∞íÏùò Î≤îÏúÑÎ•º 100Í∞úÏùò Íµ¨Í∞ÑÏúºÎ°ú ÎÇòÎàÑÏñ¥ ÎÇúÏàòÏùò ÎπàÎèÑÏàòÎ•º ÌûàÏä§ÌÜ†Í∑∏Îû®ÏúºÎ°ú Í∑∏Î¶º
        # Í∑∏ÎûòÌîÑ Íæ∏ÎØ∏Í∏∞
        axes[i, j].set_title("mean:"+str(i*10)+", std:"+str(3*(j+1)), fontsize=20)  # Îç∞Ïù¥ÌÑ∞Ïùò ÌèâÍ∑†Í∞íÍ≥º ÌëúÏ§ÄÌé∏Ï∞®Í∞íÏùÑ Ï†úÎ™©Ïóê Ï∂îÍ∞Ä
        axes[i, j].grid(axis='both', color='purple', linestyle=':', linewidth=1)  # Í≤©Ïûê ÏÑ§Ï†ï
        axes[i, j].set_facecolor('gray')  # axesÏùò Î∞îÌÉïÏÉâ ÏÑ§Ï†ï
        
plt.show()  # figureÎ•º Ï∂úÎ†•
```
