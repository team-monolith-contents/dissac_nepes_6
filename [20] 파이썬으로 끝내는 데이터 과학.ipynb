###  **1. 그래프의 배경 꾸미기**

> 그래프를 그리는 여러 함수에는 여러가지의 배경 꾸미기 기능이 숨어있습니다.  
> 다음 코드를 실행하여 그래프를 얻어 확인해보고 어떤 기능이 사용됐는지 찾아봅시다.  


```python
# 예제 1: 다양한 배경 꾸미기 기능이 사용된 그래프 확인하기

import numpy as np

plt.figure(figsize=(10, 7), facecolor='pink', edgecolor='c', linewidth=2)
plt.axes(facecolor='y')

plt.title("왼쪽 제목", loc='left')
plt.title("오른쪽 제목", loc='right')
plt.title("제목, fontsize=40", fontsize=40, fontweight='bold', pad=20)

plt.xlabel("x축 이름, fontsize=20", fontsize=20, loc='right')
plt.ylabel("y축 이름", rotation='horizontal')

plt.xticks([-2, -1, 0, 1, 2], fontsize=15, fontweight='bold', color='purple')
plt.yticks([np.exp(-2), np.exp(-1), np.exp(0), np.exp(1), np.exp(2)], 
           ["exp(-2)="+str(np.exp(-2)), "exp(-1)="+str(np.exp(-1)), "exp(0)="+str(np.exp(0)), "exp(1)="+str(np.exp(1)), "exp(2)="+str(np.exp(2))]
           ,rotation=30)

plt.xlim(-2.5, 2.5)
plt.ylim(0, 8)

plt.grid(axis='x', color='purple', linestyle='--', linewidth=3)
plt.grid(axis='y', color='gray')

plt.show()
```

### **2. 그래프의 배경 크기와 색 설정하기**

> Pyplot의 그림은 두 부분으로 나뉩니다.  
> * **figure**: 배경을 의미. 제목, 축 이름, 눈금의 이름을 포함.  
> * **axes**: 좌표축을 의미. 눈금과 격자, 그래프를 포함.
>
> 다음과 같이 배경, 좌표축의 크기, 색상, 모서리 색상, 모서리 굵기를 설정할 수 있습니다.  
>
> * `plt.figure(figsize=(w, h), facecolor=C1, edgecolor=C2, linewidth=n)`  
> : 너비가 w인치, 높이가 h인치인 크기의 배경을 만듬  
> : 안쪽 색상을 C1으로 채움  
> : 모서리의 색상은 C2이고 굵기는 n(숫자)으로 그림
> * `plt.axes(facecolor=C3)`  
> : 안쪽 색상을 C3으로 채움  

```python
# 예제 2: 그래프의 배경 크기와 색, 좌표축의 색을 결정

plt.figure(figsize=(2, 3), facecolor='w', edgecolor='k', linewidth=5)  # figure를 너비 2인치, 높이 3인치, 바탕색은 흰색, 모서리는 굵기5의 검정선으로 설정
plt.axes(facecolor='gray')  # axes의 바탕색을 회색으로 설정

plt.show()  # figure를 출력
```

### **3. 그래프 제목과 축의 이름 설정하기**

> Figure는 두가지 종류의 글자를 포함합니다.  
>
> * **title**: 그래프의 제목
> * **xlabel/ylabel**: x/y축의 이름
> 
> 다음과 같이 그래프 제목, x/y축 이름의 위치, 크기, 글자 스타일을 설정할 수 있습니다.  
>
> * `plt.title("제목", fontsize=N, color=c, fontweight=weight, pad=M, loc=lacation)`  
> : "제목"이라는 제목을 글자크기 N(숫자), 글자색상은 c(색깔이름), 글자굵기를 weight(숫자 혹은 **normal/bold**)로 설정  
> : 제목과 axes 사이의 간격을 M(숫자)포인트로 설정  
> : 제목의 위치를 location(**left(좌)/center(중앙)/right(우)**)으로 설정  
>
> * `plt.xlabel("x축", fontsize=P, color=c, fontweight=weight, loc=location, rotation=direction)`  
> : "x축"이라는 축이름을 글자크기 P(숫자), 글자색상은 c(색깔이름), 글자굵기를 weight(숫자 혹은 **normal/bold**)로 설정  
> : 축이름의 위치를 location(**left(좌)/center(중앙)/right(우)**)으로 설정  
> : 축이름의 글씨방향을 direction(**horizontal(가로방향)/vertical(세로방향)**)으로 설정  
>
> * `plt.ylabel("y축", fontsize=P, color=c, fontweight=weight, loc=location, rotation=direction)`  
> : loc 인수 외에는 xlabel과 같음  
> : 축이름의 위치를 location(**top(위)/center(중앙)/bottom(아래)**)으로 설정  


```python
# 예제 3: 그래프 제목과 축이름의 내용, 크기, 스타일, 위치를 결정

## (1) 그래프 제목 설정
# "왼쪽 제목"을 왼쪽에 글자크기 20, 빨간색으로 추가
plt.title("왼쪽 제목", loc='left', fontsize=20, color='r')

# "오른쪽 제목"을 오른쪽에 글자크기 20, normal 굵기로 추가
plt.title("오른쪽 제목", loc='right', fontsize=20, fontweight='normal')  

# "중앙 제목"을 중앙에 글자크기 20, bold 굵기로 추가 후 axes와의 간격을 20포인트로 설정
plt.title("중앙 제목", fontsize=20, fontweight='bold', pad=20)  


## (2) 그래프 축 이름 설정
# x축의 이름 "x축 이름"을 오른쪽에 글자크기 15, 파란색, bold 굵기로 추가
plt.xlabel("x축 이름", loc='right', fontsize=15, color='b', fontweight='bold')  

# y축의 이름 "y축 이름"을 위쪽에 글자크기 15, bold 굵기로 가로방향 글자로 추가
plt.ylabel("y축 이름", loc='top', fontsize=15, fontweight='bold', rotation='horizontal')  

plt.show()  # figure를 출력
```

### **4. 축의 눈금 설정하기**

> Axes는 아래에 x축, 왼쪽에 y축을 가집니다.   
>
> * **xticks/yticks**: x/y축의 눈금 이름
> * **xlim/ylim**: x/y축의 범위
> 
> 다음과 같이 각 축의 눈금을 설정할 수 있고, 보여지는 축의 범위를 설정할 수 있습니다.  
>
> * `plt.xticks/yticks(num_data, label_data, fontsize=N, color=c, fontweight=weight, rotation=angle)`  
> : num_data의 숫자들의 간격에 맞추어 눈금을 지정  
> : label_data가 있는 경우 위의 눈금 이름을 label_data로 대체    
> : 눈금 이름을 글자크기 N(숫자), 글자색상은 c(색깔이름), 글자굵기를 weight(숫자 혹은 normal/bold)로 설정  
> : 눈금 이름을 angle(숫자) 각도만큼 반시계방향으로 회전  
>
> * `plt.xlim/ylim(minumum, maximum)`  
> : axes에 보여지는 축의 범위를 minumum부터 maximum까지로 설정 

```python
# 예제 4: 축의 눈금 위치와 이름을 결정

## (1) 축의 눈금 설정
# x축의 눈금을 일정한 간격으로 지정하고 눈금 이름을 글자크기 15, bold굵기, 보락색으로 추가
plt.xticks([-2, -1, 0, 1, 2], fontsize=15, fontweight='bold', color='purple')  

# y축의 눈금을 다음과 같이 추가
plt.yticks([np.exp(-2), np.exp(-1), np.exp(0), np.exp(1), np.exp(2)],  # y축의 눈금을 exponential 함수를 이용해 다양한 간격으로 지정
           ["exp(-2)="+str(np.exp(-2)), "exp(-1)="+str(np.exp(-1)), "exp(0)="+str(np.exp(0)), "exp(1)="+str(np.exp(1)), "exp(2)="+str(np.exp(2))]  # 눈금 이름을 'exp(n)=실제값'으로 지정
           ,rotation=30)  # 눈금 이름을 반시계방향으로 30도 회전


## (2) 축의 범위 설정
plt.xlim(-2.5, 2.5)  # x축을 -2.5부터 2.5까지 보여지도록 설정
plt.ylim(0, 8)  # y축을 0부터 8까지 보여지도록 설정

plt.show()  # figure를 출력
```

### **1. 꺾은선 그래프에서 스타일 설정 배우기**

> plot 함수로 그려지는 꺾은선 그래프는 **line/marker** 두가지 요소를 포함합니다.  
> 각 요소를 지정할 수 있는 다양한 스타일을 확인하고 그래프를 그려봅시다.
>
> 또한 Axes 속에 그려지는 범례를 설정하는 방법도 알아봅시다. 

**(1) 선 스타일 설정하기**

> 다음과 같이 line의 색상, 스타일, 굵기를 지정할 수 있습니다. 
> * **line color** : `color` 인수에 이름을 넣어 색상 지정    
> |이름|색상|이름|색상|이름|색상|이름|색상|이름|색상|이름|색상|이름|색상|
> |:---:|---|:---:|---|:---:|---|:---:|---|:---:|---|:---:|---|:---:|---|
|**k**|검정|**r**|빨강|**g**|초록|**b**|파랑|**gray**|회색|**pink**|분홍|**purple**|보라|
|**w**|하양|**c**|사이언|**m**|마젠타|**y**|노랑|**orange**|주황|**lime**|연두|**navy**|남색|
>* **line style** : `linestyle` 인수에 마크를 넣어 선스타일 지정
> |마크|선스타일|이름|
|:---:|---|---|
|**-**|실선|solid|
|**--**|----------|dashed|
|**-.**|-.-.-.-.-.-|dashdot|
|**:**|.............|dotted|
>* **line width** : `linewidth`인수에 숫자를 넣어 굵기를 지정

```python
# 예제 1: 꺾은선 그래프의 선 스타일 설정하기

## (1) 그래프의 배경 설정
plt.figure(figsize=(10, 6), facecolor='pink')  # 너비 10인치, 높이 6인치의 바탕색이 분홍인 figure를 생성 
plt.title("꺾은선 그래프의 다양한 스타일", fontsize=25, fontweight='bold', pad=15)  # 제목을 생성
plt.xlabel("x축", fontsize=20)  # x축 이름을 생성
plt.ylabel("y축", fontsize=20, labelpad=30, rotation='horizontal')  # y축 이름을 설정하고 가로 방향으로 생성
plt.xlim(-0.5, 9.5)  # x축 범위를 -0.5부터 9.5로 설정
plt.ylim(-0.5, 9.5)  # y축 범위를 -0.5부터 9.5로 설정
plt.grid(axis='y', color='purple', linestyle=':')  # y축 눈금 격자를 보라색 점선으로 생성


## (2) 그래프의 스타일을 설정
plt.plot(range(10), range(10), color='b', linestyle='-', linewidth=2, label='blue solid')  # 파란색 solid 선그래프
plt.plot(range(6), [1, 2.5, 4, 5.5, 7, 8.5], color='g', linestyle='-.', linewidth=2, label='green dashdot')  # 초록색 dashdot 선그래프
plt.plot([2, 4, 6, 8], [0.5, 1.5, 2.5, 3.5], color='r', linestyle=':', linewidth=2, label='red dotted')  # 빨간색 dotted 선그래프
plt.plot(range(10), [9, 5, 3.5, 2.5, 2, 1.8, 1.5, 1.3, 1.2, 1.15], color='orange', linestyle='--', linewidth=3, label='orange dashed')  # 주황색 dashed 선그래프

plt.show()  # figure를 출력
```

**(2) 선 위의 점 스타일 설정하기**

> 다음과 같이 marker의 모양과 크기를 지정할 수 있습니다.  
>
> * **marker** : `marker`인수에 마크를 넣어 모양 지정  
> |마크|모양|마크|모양|마크|모양|마크|모양|마크|모양|
> |:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|:---:|
> |**o**||**^**||**v**||**s**||**P**||
|*||**D**||**+**||**x**||**X**||
> 
> * **marker size** : `markersize` 인수에 숫자를 넣어 크기를 지정 (선 위의 모든 데이터 marker가 같은 크기)

### **2. 그래프 세로방향으로 나열하기**

>
> `figure, axes = plt.subplots(N, figsize=(w, h))`   
> : N개의 axes가 세로방향으로 나열되어 있고 너비가 w 인치, 높이가 h 인치인 figure와 axes리스트를 반환  
>  
> 위에서 얻은 axes는 여러개의 그래프(axes)를 포함하는 리스트입니다.  
> 따라서 각 그래프(axes)를 꾸밀 때에는 `axes[i]` 와 같이 세로방향 순서를 이용해 인덱싱하여 불러와 사용합니다.

> 다음과 같이 `subplots` 함수에 figure에 추가할 axes에 개수를 넣어 세로방향으로 나열된 axes와 figure를 얻을 수 있습니다.

> * **edge color** : `edgecolor` 인수에 이름을 넣어 모서리 색 지정

>
> 또한, 여러 axes들의 축을 공유할 것인지 설정할 수 있습니다.  
>
> * **share x** : `sharex` 인수에 True/False를 넣어 x축을 공유할 것인지 지정

>
> `figure.suptitle("전체 제목")` : "전체 제목"이라는 제목을 추가
> * **color** : `color` 인수에 이름을 넣어 글자색 지정
> * **fontsize** : `fontsize` 인수에 숫자를 넣어 글자크기 지정
> * **fontweight** : `fontweight` 인수에 숫자 혹은 normal/bold를 넣어 글자굵기 지정

```python
# 예제 2: 여러 axes가 세로방향으로 나열된 figure 꾸미기

# 2개의 axes가 세로방향으로 나열되어 있고, 너비 10인치 높이 6인치인 figure와 axes리스트를 반환
# figure의 바탕색은 노랑, 모서리 색은 검정, 모서리 굵기는 2로 설정
# axes들의 x축을 공유
fig, axes = plt.subplots(2, figsize=(10, 6), facecolor='yellow', edgecolor='k', linewidth=2, sharey=True)

# 글자크기 20, bold 굵기의 전체 제목 추가
fig.suptitle("그래프 여러개 나열하기", fontsize=20, fontweight='bold')

# 그래프 그리기
for i in range(2):
    axes[i].set_title("axes["+str(i)+"]")

# axes[0]의 x, y축 범위 설정
axes[0].set_xlim(-1, 1)
axes[0].set_ylim(-1, 1)

plt.show()  # figure를 출력
```

> * **그래프 꾸미기** : 제목, 눈금, 격자 등을 설정하는 함수는 이름이 약간 다르니 주의해야 합니다.  
>
>> `axes[i].set_title()` : (i-1)번째 그래프의 제목을 설정 (title 함수와 인수가 같음)  
>>
>> `axes[i].set_xticks()` : (i-1)번째 그래프의 x축 눈금을 설정 (xticks 함수와 인수가 같음)  
>> `axes[i].set_yticks()` : (i-1)번째 그래프의 y축 눈금을 설정 (yticks 함수와 인수가 같음)

>
> 다음 예제를 실행하여 각 그래프(axes)의 번호를 확인하여 봅시다. 

```python
# 예제 6: 여러 axes가 2차원으로 나열된 figure 확인하기

fig, axes = plt.subplots(2, 3, figsize=(10, 6))  # 2개의 axes가 세로방향으로 나열되어 있고, 너비 10인치 높이 6인치인 figure와 axes리스트를 반환

fig.suptitle("그래프 여러개 나열하기", fontsize=20)

for i in range(2):
    for j in range(3):
        axes[i, j].set_title("axes["+str(i)+", "+str(j)+"]")
plt.show()
```

**(2) figure와 여러개 axes 꾸미기**

> 2차원으로 나열된 그래프를 꾸미는 방법도 다르지 않습니다.  
>
> 이번에는 for문을 이용하여 각 그래프에 접근하고 히스토그램을 그려보겠습니다. 

```python
# 예제 7: figure와 여러 axes 꾸미기, for문 이용하기

## (1) figure 생성
# axes가 가로방향 2줄 세로방향 3줄로 총 6개 나열되어 있고, 너비 16인치 높이 10인치의 figure를 반환
# 또한 x축과 y축이 같은 axes리스트를 반환
fig, axes = plt.subplots(2, 3, figsize=(16, 10), sharex=True, sharey=True)

## (2) figure 제목 설정
fig.suptitle('여러 그래프 2차원으로 나열하기', fontsize=20, fontweight='bold')

## (3) 그래프 그리기 및 꾸미기
for i in range(2):
    for j in range(3):
        # 데이터
        numbers = i*10 + 3*(j+1)*np.random.standard_normal(10000)  # 표준정규분포를 확률밀도함수로 하여 난수를 10000개 생성하고 좌표에 따라 특정한 수를 곱하거나 더함
        # 그래프 그리기
        axes[i, j].hist(numbers, bins=100, density=True, color='lime')  # 값의 범위를 100개의 구간으로 나누어 난수의 빈도수를 히스토그램으로 그림
        # 그래프 꾸미기
        axes[i, j].set_title("mean:"+str(i*10)+", std:"+str(3*(j+1)), fontsize=20)  # 데이터의 평균값과 표준편차값을 제목에 추가
        axes[i, j].grid(axis='both', color='purple', linestyle=':', linewidth=1)  # 격자 설정
        axes[i, j].set_facecolor('gray')  # axes의 바탕색 설정
        
plt.show()  # figure를 출력
```
